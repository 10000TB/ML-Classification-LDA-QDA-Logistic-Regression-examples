```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def train(X,T,lamb=0):
    means = X.mean(axis=0)
    stds= X.std(axis=0)
    Xs = (X - means)/stds
    Xs1= np.hstack((np.ones((Xs.shape[0],1)), Xs))
    w = np.linalg.lstsq(np.dot(Xs1.T,Xs1),np.dot(Xs1.T,T))[0]
    model = {'means':means,'stds':stds,'w':w}
    return model
```

```python
def use(X,model):
    Xs = (X-model['means'])/model['stds']
    Xs1 = np.hstack((np.ones((X.shape[0],1)), Xs))
    return np.dot( Xs1, model['w'])
```

```python
def rmse(predict,T):
    rmseValue = np.sqrt(np.mean((predict-T)**2))
    return rmseValue
```

```python
def partitionKFolds(X,T,nFolds,shuffle=False,nPartitions=3):
    '''Usage: for Xtrain,Ttrain,Xval,Tval,Xtest,Ttext in partitionKFolds(X,T,5):'''
    # Randomly arrange row indices
    rowIndices = np.arange(X.shape[0])
    if shuffle:
        np.random.shuffle(rowIndices)
    # Calculate number of samples in each of the nFolds folds
    nSamples = X.shape[0]
    nEach = int(nSamples / nFolds)
    if nEach == 0:
        raise ValueError("partitionKFolds: Number of samples in each fold is 0.")
    # Calculate the starting and stopping row index for each fold.
    # Store in startsStops as list of (start,stop) pairs
    starts = np.arange(0,nEach*nFolds,nEach)
    stops = starts + nEach
    stops[-1] = nSamples
    startsStops = list(zip(starts,stops))
    # Repeat with testFold taking each single fold, one at a time
    for testFold in range(nFolds):
        if nPartitions == 3:
            # Repeat with validateFold taking each single fold, except for the testFold
            for validateFold in range(nFolds):
                if testFold == validateFold:
                    continue
                # trainFolds are all remaining folds, after selecting test and validate folds
                trainFolds = np.setdiff1d(range(nFolds), [testFold,validateFold])
                # Construct Xtrain and Ttrain by collecting rows for all trainFolds
                rows = []
                for tf in trainFolds:
                    a,b = startsStops[tf]                
                    rows += rowIndices[a:b].tolist()
                Xtrain = X[rows,:]
                Ttrain = T[rows,:]
                # Construct Xvalidate and Tvalidate
                a,b = startsStops[validateFold]
                rows = rowIndices[a:b]
                Xvalidate = X[rows,:]
                Tvalidate = T[rows,:]
                # Construct Xtest and Ttest
                a,b = startsStops[testFold]
                rows = rowIndices[a:b]
                Xtest = X[rows,:]
                Ttest = T[rows,:]
                # Return partition matrices, then suspend until called again.
                yield Xtrain,Ttrain,Xvalidate,Tvalidate,Xtest,Ttest,testFold
        else:
            # trainFolds are all remaining folds, after selecting test and validate folds
            trainFolds = np.setdiff1d(range(nFolds), [testFold])
            # Construct Xtrain and Ttrain by collecting rows for all trainFolds
            rows = []
            for tf in trainFolds:
                a,b = startsStops[tf]                
                rows += rowIndices[a:b].tolist()
            Xtrain = X[rows,:]
            Ttrain = T[rows,:]
            # Construct Xtest and Ttest
            a,b = startsStops[testFold]
            rows = rowIndices[a:b]
            Xtest = X[rows,:]
            Ttest = T[rows,:]
            # Return partition matrices, then suspend until called again.
            yield Xtrain,Ttrain,Xtest,Ttest,testFold
```

```python
def makeMPGData(filename='auto-mpg.data'):
    def missingIsNan(s):
        return np.nan if s == b'?' else float(s)
    data = np.loadtxt(filename, usecols=range(8), converters={3: missingIsNan})
    print("Read",data.shape[0],"rows and",data.shape[1],"columns from",filename)
    goodRowsMask = np.isnan(data).sum(axis=1) == 0
    data = data[goodRowsMask,:]
    print("After removing rows containing question marks, data has",data.shape[0],"rows and",data.shape[1],"columns.")
    X = data[:,1:]
    T = data[:,0:1]
    Xnames =  ['cylinders','displacement','horsepower','weight','acceleration','year','origin']
    Tname = 'mpg'
    return X,T,Xnames,Tname
```

```python
def makeFeatures(X,maxPower):
    newXPowered = []
    #append in specific power of X into newXPowered
    for power in range(1, maxPower+1):
        newXPowered.append(X**power)
    #stack each each powered X together
    return np.hstack(newXPowered)
```

```python
def findBestPower(X,T,nFolds,maxPowers):
    results=[]
    #calculate the matrix
    for currentMaxPower in maxPowers:
        XnewFeatures = makeFeatures(X, currentMaxPower)
        for Xtrain,Ttrain,Xtest,Ttest,testFold in partitionKFolds(XnewFeatures,T,nFolds,shuffle=False,nPartitions=2):
            currentModel = train(Xtrain,Ttrain,lamb=0)
            #print currentModel
            trainRMSE = rmse(Ttrain,use(Xtrain,currentModel))
            testRMSE = rmse(Ttest,use(Xtest,currentModel))               
            newResult = [currentMaxPower, trainRMSE, testRMSE]
            results.append(newResult)
    
    theMatrix = meansAfterGroupedByFirstColumn(np.array(results))
    
    bestMaxPower=1
    minTestRMSE=0
    firstTime=True
    #loop through all records and find the record with smallest test RMSE, and return the power in that record
    for power,trainRMSE,testRMSE in theMatrix:
        if firstTime==True:
            bestMaxPower=power
            minTestRMSE=testPower
            firstTime=False
        else:
            if minTestRMSE>testRMSE:
                minTestRMSE=testRMSE
                bestMaxPower=power

    return bestMaxPower,theMatrix
```

```python
def meansAfterGroupedByFirstColumn(data):
    result = []
    uniqs = np.unique(data[:,0])
    for x in uniqs:
        forx = data[data[:,0] == x, 1:]
        result.append([x] + np.mean(forx,0).tolist())
    return np.array(result)
```

###RMSE: ${rmse(predict,T):}$

In order to judge the precision of the predict compared to the target, we use the root-mean-square deviation(RMSE), the definion of RMSE is:

$$\begin{align*}
RMSD=\sqrt{\frac{\sum_{i=1}^n (x_i - y_i\ )^2}{n}}
\end{align*}$$    

Given $x_i$, $y_i $ as each corresponding value in two data sets, and $n$ as the total number of balues in each of them In this notebook, the rmse(predict,T), which is $$rmseValue = np.sqrt(np.mean((predict-T)**2))$$   takes two inputs, and both of them are n by 1. The $(predict-T)**2$ corresponds to $\sum_{i=1}^n (x_i-y_i)$, and  $\begin{align*}
\sqrt{\frac{\sum_{i=1}^n (x_i - y_i\ )^2}{n}}
\end{align*}$ corresponds to $np.sqrt(np.mean((predict-T)**2))$

###Weight(s) $W$

To find the best value for $\wv$, we take the derivative the the sum of squared error objective, set it equal to 0 and solve for $\wv$. 

$$
\begin{align*}
\frac{\partial \sum_{n=1}^N (\tv_n - g(\xv_n;\wv))^2}{\partial \wv} 
&= -2 \sum_{n=1}^N (\tv_n - \xv_n^T \wv) \xv_n^T
\end{align*}
$$

It is pretty beneficial to express the $\xv_n$ and $t_n$ samples as matrices as follows:
$$
\begin{align*}
-2 \sum_{n=1}^N (\tv_n - \xv_n^T \wv) \xv_n^T= -2 \Xv^T (\Tv - \Xv \wv)
\end{align*}
$$

Then, the sum can be performed with a dot product, Check the sizes and shapes of each matrix in the last equation above.

Now we can set this equal to zero and solve for $\wv$.

$$
\begin{align*}
-2 \Xv^T (\Tv - \Xv \wv) &= 0\\
\Xv^T (\Tv - \Xv \wv) &= 0\\
\Xv^T \Tv &= \Xv^T \Xv \wv\\
\wv &= (\Xv^T \Xv)^{-1} \Xv^T \Tv
\end{align*}
$$

###Sampels:$X$

However, here the model we are building is no-linear, the features we make by $makeFeatures(X,maxPower)$ for training is as follows:

$$
  \begin{align*}
    \Xv &= \begin{bmatrix} x_{0,0}+x_{0,0}^2+x_{0,0}^3\dots+x_{0,0}^{maxPower},x_{0,1}+x_{0,1}^2+x_{0,1}^3\dots+x_{0,1}^{maxPower},\dots,x_{0,N}+x_{0,N}^2+x_{0,N}^3\dots+x_{0,N}^{maxPower} \\ x_{1,0} \dots x_{1,N} \\ x_{2,0} \dots x_{2,N} \\ \vdots \\ x_{D,0} \dots x_{D,N}  \end{bmatrix}
  \end{align*}
$$

So in  $makeFeatures(X,maxPower)$, I iterate through all powers before $maxpower$(inclusive) and append in the powered value and stack them together vertically.
